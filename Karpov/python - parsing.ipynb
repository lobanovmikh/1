```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('https://stepik.org/media/attachments/course/72204/tables.xlsx')
```

```python
df
```

```python
df = df.dropna(how='all') # Удаление строк, полностью состоящих из NaN
df.reset_index(drop=True, inplace=True) # Ресет индексов
```

```python
df.isna().sum() # Cумма пустых значений в колонках
```

```python
part_df = df.loc[df['Unnamed: 0'].notna(), :].copy()  # Отделение сдвинутого массива по колонке Unnamed: 0 - есть значение
```

```python
part_df
```

```python
part_df.drop(columns=['company48'], inplace=True) # Удаление пустой колонки
```

```python
part_df.columns = part_df.loc[18] # Колонки из первой строки - в колонки датафрейма
```

```python
part_df.drop(index=18, inplace=True) # Удаление лишней строки
```

```python
part_df
```

```python
df = df.loc[df['Unnamed: 0'].isna()] # Отделение другой части
```

```python
df.isna().sum() # Пустые значения в колонках
```

```python
df.drop(columns = ['Unnamed: 0'], inplace=True) # Удаление первой пустой колонки
```

```python
df
```

```python
columns = []    # Отбор индексов строк, которые не могут перевестись во float, т.к. в них тексты (через исключения)

for i, row in df.iterrows():
    try:
        row.astype('float')
    except:
        columns.append(i)
        
columns        
```

```python
df.loc[columns] # Проверка таких строк
```

```python
df.drop(index=columns, inplace=True) # Удаление таких строк
```

```python
df = pd.concat((df, part_df)) # Cклейка двух датафреймов
```

```python
df.fillna(0, inplace=True) # Заполнение пустых значений 0
```
