```python
import pandas as pd
import numpy as np
import datetime as DT
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
%matplotlib inline

register_matplotlib_converters()

courses = pd.read_csv('courses.csv', encoding='utf8')
del courses['Unnamed: 0'] #удаление лишнего столбца
courses.columns = ['course_id', 'course_title', 'course_field'] #переименование столбцов

course_contents = pd.read_csv('course_contents.csv', encoding='utf8')

progress = pd.read_csv('progresses.csv', encoding='utf8')
progress.columns = ['progress_id', 'student_id', 'course_id'] #переименование столбцов

progress_phases = pd.read_csv('progress_phases.csv', encoding='utf8')

progress_sum = pd.merge(progress_phases, progress, on='progress_id').fillna(0) #слияние массивов и заполнением пропущенных ячеек нулями

courses_sum = pd.merge(courses, course_contents, on='course_id').fillna(0)

main = pd.merge(progress_sum, courses_sum, on=['course_id', 'module_number', 'lesson_number'],  how='left').fillna(0) #финальный датафрейм для работы 


#1 общее количество курсов в датасете
main_courses = main.groupby('course_id').nunique() #количество уникальных курсов в датафрейме
courses_count = main_courses.shape[0] #подсчет строк
print('Общее количество курсов в датасете - ', courses_count)

#2 количество модулей на каждом курсе
main2 = main.groupby('course_title').nunique()
print('\n============ Количество модулей на каждом курсе =============')
print(main2.loc[ : , ['module_title']]) #вывод количества уникальных значений модулей в каждом курсе

#3 количество уроков в каждом модуле на каждом курсе

main3 = main.groupby(['course_title', 'module_title']).nunique() #группировка по модулям для каждого курса с уникальными значениями
main3 = main3.loc[ : ,['lesson_title']]
pd.set_option('display.max_rows', main3.shape[0]+1)
print('\n============ Количество уроков в каждом модуле на каждом курсе =============')
print(main3)

#4 медианное количество уроков в модуле на каждом курсе
main4 = main.groupby(['course_title', 'module_title']).nunique()
main4 = main3.loc[ : ,['lesson_title']]
main4.reset_index(inplace=True)
main4 = main4.groupby(['course_title']).agg({'lesson_title':'median'}) #медианное количество уроков в модуле для каждого курса
print('\n============ Медианное количество уроков в модуле на каждом курсе =============')
print(main4)


#5 количество учеников на каждом курсе
main5 = main.groupby('course_title').nunique() 
main5 = main5.loc[ : , ['student_id']] #количество уникальных студентов на каждом курсе
print('\n============ Количество учеников на каждом курсе =============')
print(main5)


#6 минимальный, максимальный, средний, медианный возраст студентов

students = pd.read_csv('students.csv', encoding='utf8')
del students['id_'] #удаляем лишний столбец
students.columns = ['student_id', 'student_city', 'student_birthday'] #переименование столбцов


students['student_birthday'] =  pd.to_datetime(students['student_birthday'], errors='coerce') #заполнение пустыми значениями ячеек, которые не соответствуют дате (ошибки)
students.dropna(inplace=True) #удаление строк с ошибочными значениями
students['age'] = DT.datetime.now().year - students.student_birthday.dt.year #разница между текущим годом и годом рождения студента - возраст студентов



students.drop(students.index[students.age < 14], inplace=True) #удаление строк для студентов меньше 14 лет (считаем, что минимальный возраст студента - 14)
students.drop(students.index[students.age > 70], inplace=True) #удаление строк для студентов больше 70 лет (считаем, что максимальный возраст студента - 70)
print('\nМинимальный возраст студентов - ', students.age.min())
print('Максимальный возраст студентов - ', students.age.max())
print('Средний возраст студентов - ', students.age.mean())
print('Медианный возраст студентов - ', students.age.median()) #расчет показателей для возраста студентов

#7 минимальный, максимальный, средний, медианный возраст студентов на каждом курсе

mainst = pd.merge(main, students, on=['student_id'],  how='left').dropna() #слияние исходного фрейма с инф. по студентам 

#находим вышеказанные показатели

main_st1 = mainst.groupby(['course_title']).agg('min')
main_st1 = main_st1.loc[ : ,['age']]
print('\n============ Минимальный возраст студентов на каждом курсе =============')
print(main_st1)

main_st2 = mainst.groupby(['course_title']).agg('max')
main_st2 = main_st2.loc[ : ,['age']]
print('\n============ Максимальный возраст студентов на каждом курсе =============')
print(main_st2)

main_st3 = mainst.groupby(['course_title']).agg('mean')
main_st3 = main_st3.loc[ : ,['age']]
print('\n============ Средний возраст студентов на каждом курсе =============')
print(main_st3)

main_st4 = mainst.groupby(['course_title']).agg('median')
main_st4 = main_st4.loc[ : ,['age']]
print('\n============ Медианный возраст студентов на каждом курсе =============')
print(main_st4)

# п.2 Постройте bar-chart, отражающий количество студентов на каждом курсе. Ticks нужно развернуть так, чтобы они были читаемы
main5 = pd.to_numeric(main5['student_id'], downcast='integer') #количество студентов переводим в целый тип
fig, subplot = plt.subplots()
subplot.bar(main5.index.values, main5.values)
plt.xticks(rotation=45, ha="right") #поворот названий для x
print('\n============ Bar-chart количества учеников на каждом курсе =============')
plt.show
```

```python
#1. Рассчитать прирост студентов на каждом курсе в каждом месяце за всю историю (каждый месяц в диапазоне от марта 2016 до июля 2019 включительно)
#2 На основании первого пункта построить line-graph с приростом студентов в каждом месяце для каждого курса. 15 графиков.

main = pd.merge(progress_sum, courses_sum, on=['course_id', 'module_number', 'lesson_number'],  how='left').fillna(0) #обновляем исходный массив для работы


main.start_date = pd.to_datetime(main.start_date)
main.finish_date = pd.to_datetime(main.finish_date) #перевод столбцов в формат даты

main2_1 = main.query('is_homework == True').groupby(['course_title','student_id']).agg({'start_date':'min'}) #считаем что начало учебы - дата первой д/з (наличие д/з и мин.дата)
main2_1 = main2_1.sort_values(by='start_date').reset_index() #сбрасываем индексацию

main2_1['year_month'] = main2_1['start_date'].apply(lambda x:x.strftime('%Y-%m')) #создаем столбец, учитывающий только год и месяц старта

main_st = main2_1.groupby(['course_title', 'year_month']).agg({'student_id':'count'}).sort_values(by='year_month') #считаем прирост студентов в каждом месяце для каждого курса



date_range = pd.date_range('2016-03','2019-07',freq='M') #создание отдельного фрейма дат для дополнения до одинакого диапазона на каждом курсе
course_list = main.course_title.unique()
main_copy = main2_1.copy()

for course in course_list: #двойной цикл по названию курса и дате для полученного фрейма курсов с одинаковым диапазоном дат
    for date in date_range:
        date = date.strftime('%Y-%m')
        if course in main_copy.query('year_month == @date')['course_title'].values:
            continue
        else:
            main_copy = main_copy.append({'year_month':date, 'course_title':course}, ignore_index=True)




# для удобства для каждого курса создаем отдельный массив и по нему строим графики (15 штук)

for i in range(courses_count):
    main1 = list(main_copy.groupby(['course_title']))[i][1] 
    current_course = list(main1['course_title'].unique())
    main1 = main1.drop( ['course_title', 'start_date'], axis='columns').reset_index()
    del main1['index']
    main1 = main1.groupby(['year_month']).agg({'student_id':'count'})
    print('\n=======', current_course[0], '======')
    print(main1)
    fig, subplot = plt.subplots()
    x_ticks = main1.index.values
    y_ticks = main1['student_id'].values
    subplot.plot(x_ticks,y_ticks)
    plt.xticks(rotation=90, ha="right")
    plt.tick_params(axis='x', labelsize=7)
    plt.title(u'Количество студентов по месяцам на курсе ' + current_course[0])
    plt.xlabel('\nМесяц года')
    plt.ylabel(u'Количество студентов')
    plt.show
```

## 4. Расчет конверсии

1. **(p)** Посчитать конверсию перехода студентов из одного модуля в другой на каждом курсе. Формула: отношение количества студентов, приступивших к выполнению домашнего задания в этом модуле (если дз в модуле несколько, то считать по первому дз в модуле), к количеству студентов, сдавших задание в предыдущем модуле (если дз в модуле несколько, то считать по последнему дз в модуле).

3. **(m)** Постройте горизонтальный (столбцы должны располагаться горизонтально) bar-chart, отражающий конверсию перехода студентов из одного модуля в другой на каждом курсе. 15 графиков. Графики должны иметь заголовки. Ticks должны содержать номер и название модуля. Цвет столбцов графиков должен содержать информацию о сфере, к которой относится курс (нужна легенда). Прозрачность должна стоять на отметке 0.1. На графики должна быть нанесена линия медианы конверсии для каждого курса. У медианы должен быть свой цвет. Рамок у графиков быть не должно ⭐

```python
main = pd.merge(progress_sum, courses_sum, on=['course_id', 'module_number', 'lesson_number'],  how='left').fillna(0) #обновляем исходный датафрейм

main4_1_1 = main.query('is_homework == True')
main4_1_1 = main4_1_1.groupby(['course_title', 'module_title', 'module_number', 'student_id']).agg({'start_date':'min'}).reset_index()
main4_1_1 = main4_1_1.groupby(['course_title', 'module_number', 'module_title']).agg({'student_id':'nunique'})
main4_1_1.rename(columns={'student_id':'st_start'}, inplace=True) #получаем число студентов, приступивших к выполнению д/з в модуле

main4_1_2 = main.query('is_homework == True and status ==@d')
main4_1_2 = main4_1_2.groupby(['course_title', 'module_title', 'module_number', 'student_id']).agg({'finish_date':'max'}).reset_index()
main4_1_2 = main4_1_2.groupby(['course_title', 'module_number', 'module_title']).agg({'student_id':'nunique'}).reset_index() 
main4_1_2.rename(columns={'student_id':'st_finish'}, inplace=True) #получаем число студентов, завершивших последнее задание в модуле
main4_1_2['st_finish + 1'] = main4_1_2.groupby(['course_title'])['st_finish'].transform(lambda x:x.shift()) #сдвигаем строки числа финиша вниз на 1 строку для каждого курса для расчета конверсии, с учетом того, что для первых модулей курса конверсия не считается, т.к. отсутствует предыдущий модуль (порядок прохождения курсов может быть различным)



main4_1 = pd.merge(main4_1_1, main4_1_2, on=['course_title', 'module_number', 'module_title'],  how='left').dropna(0) #соединяем исходный массив с числом приступивших студентов и завершивших предыдущий модуль
pd.set_option('display.max_rows', main4_1.shape[0]+1) #полное отображение массива для наглядности
main4_1['st_start'] = main4_1['st_start'].astype(float)
main4_1['conversion'] = main4_1['st_start'] / main4_1['st_finish + 1'] #расчет конверсии для каждого модуля (первые модули для каждого курса были исключены)

main4_1


```

```python
main4_2 = list(main4_1.groupby(['course_title']))

#построение bar-chartов для каждого курса, по оси x - модули, по оси y - значение конверсии для модуля

for i in range(courses_count):
    fig, subplot = plt.subplots()
    current_course_list = main4_2[i][1]
    x_ticks =  current_course_list.loc[ : , 'module_title'].values
    y_ticks = current_course_list.loc[ : , 'conversion'].values
    subplot.bar(x_ticks,y_ticks)
    plt.xticks(rotation=35, ha="right")
    plt.title(u'Конверсия перехода студентов из одного модуля в другой на курсе ' + main4_2[i][0])
    plt.show
```
